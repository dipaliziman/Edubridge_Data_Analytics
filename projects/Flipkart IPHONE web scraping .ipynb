```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
uclient = urlopen(url)
page_html = uclient.read()
uclient.close()
```

```python
page_soup = soup(page_html,"html.parser")
```

```python
container = page_soup.find_all("div",{"class":"_2kHMtA"})
```

```python
con = container[0]
print(soup.prettify(con))
```

## product name

## Adding more urls

```python
url = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
urls = []
```

```python
for var in range(1,16):
    url=["https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=%i"%(var)]
    urls.append(url[0])
```

```python
urls
```

```python
def connect(ur):
    uclient = urlopen(ur)
    page_html = uclient.read()
    uclient.close()
    
    page_soup = soup(page_html,"html.parser")
    
    container = page_soup.find_all("div",{"class":"_2kHMtA"})
    return container
    
```

```python
con = connect(urls[0])
print(soup.prettify(con[1]))
```

```python
con_urls = connect(urls[0])
```

```python
for i in range(0,15):
    con = connect(urls[i])
    for container in con:
        product_name = container.div.img["alt"]
        print(product_name)
```
