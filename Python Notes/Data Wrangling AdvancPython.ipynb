## Merging data

- the pandas library provides a single function 'merge' as an entry point of all standard database join,operations between dataframe objects

#syntax :pd.merge(left,right, how='inner', on=None, left_on=None,right_on=None, left_index=False, right_index=False, sort=True)

     #parameter: 
     
     1. left \right :dataframe or named series.It is an Object which merges with data frame.
     
     2. how : (left,right,outer,inner default are inner) 
             
             1. left - it used only key left frame similar to sql left join
             
             2. right - it used only key right frame similar to sql right join
             
             3. outer -it uses the union of keys from both frame i.e union
             
             4. inner- it uses the intersection of keys from both frame i.e intersection
     
     3. on : it is labels or list 
     
     4. right on : it is labels or list or array like
     
     5.left index - it is talking about booleans, default False
     
     6.right index- it is talking about boolean ,defaut False 
     
     7.sort - it is boolean ,default False
     
     
     


```python
## Example of merge
import pandas as pd
# Creatig dataframe
data1= pd.DataFrame({'id':[1,2,3,4,5],
                     'name':['serena','vinice','rohit','rafael','santiago'],
                   "branch":['science','commerce','it','computer Science','arts']})

data2=pd.DataFrame({'id':[1,2,3,4,5],
                   'fees status':['paid','paid','un-paid',"paid",'un-paid']})

#merging the data frame
print(pd.merge(data1,data2,on='id'))
pd.merge(data1,data2,on='id')
```

```python
#Without using merge
left=pd.DataFrame({"id":[1,2,3,4,5],
                  'name':['arohi','aron','aiysha','amish','amar'],
                  'Course':['data analytics','java','python','machine learning','dot net']})
right=pd.DataFrame({'id':[1,2,3,4,5],
                   'name':['bharat','ben','banny','bhushan','benedict'],
                  'Course':['dot net',"data analytics",'machine learning ','java','python']})
#print(left)

#print(right)
pd.merge(left,right, on='id')
```

## Reshaping Data

-Data is manipulated according to the requirments,where new data can be added and pre-existing data can be modified


```python
import pandas as pd
information={'name':["vicky",'ronit','dipali','tanuja'],
            'age':[25,24,21,22],
            'gender':["M",'M',"F","F"]}
data=pd.DataFrame(information)
data
```

```python
#in the gender column we can reshape the data by categories into number,m=0,F=1
#Categorize Gender
data['gender']=data['gender'].map({'M':0,'F':1}).astype(float)
data
#map function returns the map object of the result after applying result of each item
#astype is method to casting pandas object to a specific datatype
#it can also convert any suitable existing column to a categorical type

```
