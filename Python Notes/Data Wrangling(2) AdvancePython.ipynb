## Wrangling data using Grouping  method

-The grouping method in data analysis is used to provide results in terms of various group taken out from large data.

-This method of pandas used to group the outset of data from the large data set.

- General Example(understanding purpose)- There is car selling company and this company have different branch of various car manufacturing company like maruti,toyota,mahindra,ford, etc and have data where different cars are sold in different yrs.So the company wants to wrangle only that data where cars are sold during the year 2010.For this problem we used another wrangling technique which is groupby()method

```python
#Example-
import pandas as pd 
#creating a data of a car
manufacturing_company={'Brand':['maruti','maruti','maruti','toyota','toyota','ford','ford','ford'],
                      "years":[2010,2011,2012,2012,2013,2014,2010,2009],
                      'sold':[6,7,9,6,5,9,8,7]}
df=pd.DataFrame(manufacturing_company)
df
```

```python
##Group it by using groupby(),group data when year =2010
grouped=df.groupby('years')
grouped.get_group(2010)
```

## Concatenating Data-

-python pandas provide various facilities of easily combining together series dataframe object.

-To concatenate pandas dataframe use pd.concat()



```python
# Creating 2 data frame
import pandas as pd
data1=pd.DataFrame({"name":['rachael','joey','chandler','ross','monica'],
                   "Course":['java','.net','data analytics','digital marketing','c++'],
                   "marks":[90,45,78,99,75]},index=['A','B',"C",'D','E'])
data2=pd.DataFrame({"name":['Berlin','tokyo','nerobi','denver','proffesor'],
                   "Course":['it','c','data analytics','digital marketing','c++'],
                   "marks":[47,45,78,99,95]},index=['F','G',"H",'I','J'])
pd.concat([data1,data2])
```
