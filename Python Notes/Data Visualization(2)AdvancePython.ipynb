## PieChart
-it is a circular diagram often used in descriptive statistics for graphical representation of quantitative variables measured 
on classes of categories.
- python matplot lib piechart display the series of data in slices or wedges and each slice is size of an item 

- syntax of matplot lib pie function  matplotlib.pyplot.pie(x, labels = None)


```python
#example - lets use different computer languages to plot a piechart
# python is 29% ,javascript- 26%,java-21%,C-17%,C++-11% famous.
#This is just a sample data
#Define Color of visualization- To make the diagram clearer the various slices are filled with different colors
```

```python
import matplotlib.pyplot as plt
```

```python
data={'python':29,'javascript':26,'java':21,'c':17,'C++':11}
colors=['orange','green','yellow','blue','cyan']
highlight=(0,0,0,0,0)
plt.pie(list(data.values()),
        explode=highlight,
        labels=data.keys(),
        colors=colors, 
        autopct='%2.1f%%',      #percentage of each slice in matplotlib pie chart
        shadow=True, 
        startangle=90)
plt.axis('equal')
plt.show()
```

```python
data={'python':29,'javascript':26,'java':21,'c':17,'C++':11}
colors=['orange','green','yellow','blue','cyan']
highlight=(0,0.2,0,0,0)
plt.pie(list(data.values()),
        explode=highlight,
        labels=data.keys(),
        colors=colors, 
        autopct='%2.1f%%',      #percentage of each slice in matplotlib pie chart
        shadow=True, 
        startangle=90)
plt.axis('equal')
plt.show()
```

```python
data={'python':29,'javascript':26,'java':21,'c':17,'C++':11}
colors=['orange','green','yellow','blue','cyan']
highlight=(0,0,0,0,0)
plt.pie(list(data.values()),
        explode=highlight,
        labels=data.keys(),
        colors=colors, 
        autopct='%2.1f%%',      #percentage of each slice in matplotlib pie chart
        shadow=True, 
        startangle=45)
plt.axis('equal')
plt.show()
```

```python
data={'python':29,'javascript':26,'java':21,'c':17,'C++':11}
colors=['orange','green','yellow','blue','cyan']
plt.subplots()
highlight=(0,0,0,0,0)
plt.pie(list(data.values()),
        explode=highlight,
        labels=data.keys(),
        colors=colors, 
        autopct='%2.1f%%',      #percentage of each slice in matplotlib pie chart
        shadow=True, 
        startangle=90)
plt.axis('equal')
plt.title("Different Computer Languages",color='red',fontsize=20)
plt.legend()
plt.show()
```

## Bar Graphs

--when  you have categorical data you can represented with a bar graph.

--A bar graph plots  a data with the help of bars which represent value on 'y' axis and category on 'x' axis.

--bar graph used bars with varying heights to show data which belongs to specific category.


```python
import matplotlib.pyplot as plt
   
Country = ['USA','Canada','Germany','UK','France']
GDP_Per_Capita = [45000,42000,52000,49000,47000]

plt.bar(Country, GDP_Per_Capita)
plt.title('Country Vs GDP Per Capita')
plt.xlabel('Country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
years=range(2000,2012)
#yield_apples=[0.895,0.918,0.919,0.926,0.895,0.784,0.648,0.845,0.542,0.742,0.349,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.789,0.348,0.945,0.852]
plt.bar(years,yield_oranges)
plt.xlabel('years')
plt.ylabel('yield')
plt.title("Crops Yields")
```

```python
# we can also stack bars on top of each other
years=range(2000,2012)
yield_apples=[0.895,0.918,0.919,0.926,0.895,0.784,0.648,0.845,0.542,0.742,0.349,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.789,0.348,0.945,0.852]
plt.bar(years,yield_apples)
plt.bar(years,yield_oranges,bottom=yield_apples)
plt.legend(['apples','oranges'])
plt.xlabel('years')
plt.ylabel('yield')
plt.title("Crops Yields")
```

## Scatterplots 
- show many points plotted in the Cartesian plane.

-Each point represents the values of two variables. One variable is chosen in the horizontal axis and
another in the vertical axis

-Scatter plot can be created using the DataFrame.plot.scatter() methods.



```python
# example

import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df.plot.scatter(x='a', y='b')
```
