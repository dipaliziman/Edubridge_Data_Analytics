# What is Data Visualisation?

            > Data Visualisation, is a field in data analysis that deals with the visual representation of data.
            > It graphically plots data an in an effective way to communicate inferences from data.
            > Using data visualisation we can get a visual summary of the data.
            > Data visualisation plays a significant role in the representation of both small and large datasets,
              but it is especially useful in large datasets.
           

# Data Visualisation in Python:
    
            > Python offers several ploting libraries namely,
                - matplotlib
                - Seaborn 
              and many other such data visualisation packages, with different features for creating informative,
              customise appealing plots to present data in the most simple and effective way.
             

# Matplotlib and Seaborn:

            > Matplotlib and Seaborn are used for data visualisation.
__Matplotlib:

           > It is used for basic graph plotting like line graphs, bar charts, etc;
           > It mainly works with datasets and arrays.
           > It acts productively with data arrays and dataframes.
           > It is more customizible and pairs with pandas and numpy for EDA (exploratory data analysis)

__Seaborn:

           > Mainly used for statistics visualization and can perform complex visualization and fewer commands.
           > It works with entire dataset.
           > Seaborn is consider more organised and functional.
           > It has more in-built themes and mainly used for statistical analysis.

```python
# lets add labels to the axes so that we can represent the axes:

import matplotlib.pyplot as plt
import seaborn as sns

years=[2011,2012,2013,2014,2015]
yield_apples=[0.895,0.918,0.919,0.926,0.895]
plt.plot(years,yield_apples)
plt.title("PLOTTING LINE GRAPH")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
```

```python
# To plot multiple datasets on the same graph, use plt.plot function in each set:


import matplotlib.pyplot as plt
import seaborn as sns

years=range(2000,2012)
yield_apples=[0.895,0.918,0.919,0.926,0.895,0.784,0.648,0.845,0.542,0.742,0.349,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.789,0.348,0.945,0.852]

plt.plot(years,yield_apples)
plt.plot(years,yield_oranges)
plt.title("PLOTTING LINE GRAPH")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
```

```python
# We can add a legend which tells us each line graph means:
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

years=range(2000,2012)
yield_apples=[0.895,0.918,0.919,0.926,0.895,0.784,0.648,0.845,0.542,0.742,0.349,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.789,0.348,0.945,0.852]

plt.plot(years,yield_apples)
plt.plot(years,yield_oranges)
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```

# Marker:

        > To show each datapoint on our graph we can highlight them with markers using marker argument.
        > Many different marker shape like circle,diamond,square,cross are provided by matplotlib.

```python
plt.plot(years,yield_apples,marker="o")
plt.plot(years,yield_oranges,marker="x")

plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```

# Figure:
    
        > We can use plt.figure function to change the size of the figure:

```python
plt.figure(figsize=(12,6))
plt.plot(years,yield_oranges,marker="o")

plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["oranges"])
```

# Using seaborn:
        
        > An easy way to make your charts look beautiful is to use some default styles from the seaborn library.
        > These can be used globally by using sns.set_style() function.
        

```python
import seaborn as sns

sns.set_style("whitegrid")
plt.plot(years,yield_apples,marker="s")
plt.plot(years,yield_oranges,marker="*")
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```

```python
# we can change the backgroud by using darkgrid:

sns.set_style("darkgrid")
plt.plot(years,yield_apples,marker="s")
plt.plot(years,yield_oranges,marker="*")
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```

```python
# we can change the backgroud by using darkgrid:

sns.set_style("darkgrid")
plt.plot(years,yield_apples,marker="s",mec="r")
plt.plot(years,yield_oranges,marker="*",mec="g")
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```
