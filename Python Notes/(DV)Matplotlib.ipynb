```python
#to plot multiple dataset on the same graph, use plt.plot function in each set
import matplotlib.pyplot as plt
import seaborn as sns

years=range(2000,2012)
yield_apple=[0.895,0.918,0.910,0.926,0.929,0.931,0.851,0.756,0.562,0.892,8.956,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.895,0.859,0.856,0.956]

plt.plot(years,yield_apple)
plt.plot(years,yield_oranges)
plt.title('ploting the graph')
plt.xlabel('years')
plt.ylabel('yield(tons per hecter)')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

years=range(2000,2012)
yield_apple=[0.895,0.918,0.910,0.926,0.929,0.931,0.851,0.756,0.562,0.892,8.956,0.456]
yield_oranges=[0.721,0.645,0.896,0.542,0.423,0.786,0.984,0.631,0.895,0.859,0.856,0.956]

plt.plot(years,yield_apple)
plt.plot(years,yield_oranges)
plt.title('crop yields')
plt.xlabel('years')
plt.ylabel('yield(tons per hecter)')
plt.legend(['apples','oranges'])
```

```python
# to show each datapoint on graph we can highlight them with marker using marker argument
plt.plot(years,yield_apple,marker="o")
plt.plot(years,yield_oranges,marker="x")

plt.title('crop yields')
plt.xlabel('years')
plt.ylabel('yield(tons per hecter)')
plt.legend(['apples','oranges'])
```

```python
#we can use plt.figure functions to change the size of figure
```

```python
plt.figure(figsize=(12,10))
plt.plot(years,yield_oranges,marker="o")


plt.title('crop yields')
plt.xlabel('years')
plt.ylabel('yield(tons per hecter)')
plt.legend(['oranges'])
```

```python
import seaborn as sns
sns.set_style("whitegrid")

plt.plot(years,yield_apple,marker="s")
plt.plot(years,yield_oranges,marker="*")
plt.title('crop yields')
plt.xlabel('years')
plt.ylabel('yield(tons per hecter)')
plt.legend(['apple','oranges'])
```

```python
sns.set_style("darkgrid")
plt.plot(years,yield_apple,marker="s")
plt.plot(years,yield_oranges,marker="*")
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```

```python
sns.set_style("darkgrid")
plt.plot(years,yield_apple,marker="s",mec="r")
plt.plot(years,yield_oranges,marker="*",mec="g")
plt.title("CROP YIELDS")
plt.xlabel("years")
plt.ylabel("yield(tons per hecter)")
plt.legend(["apples","oranges"])
```
