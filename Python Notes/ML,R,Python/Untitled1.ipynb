```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1.1,2.2,3.5,4.5,5.2,6.2]
y=[5.7,8.3,9.2,10.1,2.6,3.8]
size=np.array([200,300,100,50,60,40])
colors=np.array(['red','blue','green','orange','pink','black'])

plt.scatter(x,y,label="high income low salary",c=colors,s=size)
plt.xlabel('saving')
plt.ylabel('income')
plt.title('scatter plot')
plt.legend()
plt.show()
```

```python
x=[1.1,2.2,3.5,4.5,5.2,6.2]
y=[5.7,8.3,9.2,10.1,2.6,3.8]
size=np.array([200,300,100,50,60,40])
colors=np.array(['red','blue','green','orange','pink','black'])

plt.scatter(x,y,label="high income low salary",c=colors,s=size,alpha=0.5)
plt.xlabel('saving')
plt.ylabel('income')
plt.title('scatter plot')
plt.legend()
plt.show()
```

### pie chaart

```python
y=np.array([35,25,10,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,25,10,15])
l=['Airtel','jio','vodafone','BSnl']
c=['r','b','g','k']
plt.pie(y,labels=l,startangle=90,shadow=True,colors=c,explode=(0,0,0.2,0))
plt.legend(loc='lower left',title='ISP')
plt.show()
```

```python
y=np.array([35,25,10,15])
l=['Airtel','jio','vodafone','BSnl']
c=['r','b','g','c']

plt.pie(y,labels=l,colors=c,explode=(0,0,0.2,0))
plt.legend(loc='lower left',title='ISP')
plt.show()
```
