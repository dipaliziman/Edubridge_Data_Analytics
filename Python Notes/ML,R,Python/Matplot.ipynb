### Marker color

# set font properties for title and Lables
. The fontdict parameter in xlabel() ylabel()

```python
x=np.array([2000, 2002, 2004, 2006, 2010])
y=np.array([240, 250, 260, 270, 280])

font1 = {'family':'Arial','color':'blue','size':20}
font2 = {'family':'times new roman','color':'red','size':15}

plt.title("sales Data",fontdict = font1)
plt.xlabel("Years", fontdict = font2)
plt.ylabel("sales progress",fontdict = font2)

plt.plot(x,y)
plt.show()
```

```python
x=np.array([2000, 2002, 2004, 2006, 2010])
y=np.array([240, 250, 260, 270, 280])

plt.title("sales Data")
plt.xlabel("Years")
plt.ylabel("sales progress")

plt.plot(x,y)
plt.show()
```

### Position of the title 
. the loc parameter in title() to position the title - values are 'left','right', and 'center' Default value is center 

```python
x=np.array([2000, 2002, 2004, 2006, 2010])
y=np.array([240, 250, 260, 270, 280])

font1 = {'family':'Arial','color':'blue','size':20}
font2 = {'family':'times new roman','color':'red','size':15}

plt.title("sales Data",fontdict = font1,loc='right')
plt.xlabel("Years", fontdict = font2)
plt.ylabel("sales progress",fontdict = font2)

plt.plot(x,y)
plt.show()
```

### Add grid lines to a plot
.the grid() function 

```python
from matplotlib import style
x=[40,60,80]
y=[12,56,70]
x1=[20,40,50]
y1=[23,34,56]
plt.plot(x,y,'g',label='2019',linewidth=5)
plt.plot(x1,y1,'c',label='2020',linewidth=5)
plt.title("chart")
plt.ylabel("range")
plt.xlabel("order")
plt.legend()
plt.grid(True,color='k')
plt.show()
```

### Stack vertical bar

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodafone','idea']
y=[17,18,29,40]
y1=[20,21,22,23]
plt.figure.figsize=(5,7)
plt.bar(x,y,label="2019",width=0.5,color='r')
plt.bar(x,y1,label="2020",width=0.5,bottom=y,color='b')

plt.xlabel('ISP')
plt.ylabel('sales')
plt.title('sales by year')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodafone','idea']
y=np.array([17,18,29,40])
y1=np.array([20,21,22,23])
y2=np.array([4,5,6,25])
plt.figure.figsize=(5,7)
plt.bar(x,y,label="2019",width=0.5,color='r')
plt.bar(x,y1,label="2020",width=0.5,bottom=y,color='b')
plt.bar(x,y2,label="2021",width=0.5,bottom=y+y1,color='g')


plt.xlabel('ISP')
plt.ylabel('sales')
plt.title('sales by year')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x=['airtel','jio','vodafone','idea']
y=[17,18,29,40]
y1=[20,21,22,23]
plt.figure.figsize=(5,7)
plt.barh(x,y,label="2019",color='r')
plt.barh(x,y1,label="2020",left=y,color='b')

plt.xlabel('ISP')
plt.ylabel('sales')
plt.title('sales by year')
plt.legend()
plt.show()
```
