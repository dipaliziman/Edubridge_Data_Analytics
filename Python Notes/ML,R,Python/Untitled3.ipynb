```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\Edu bridge\EDU\DATA\breast_cancer_train.csv")
```

```python
df
```

```python
df = df.drop(df[df["bare_nuclei"]=="?"].index,axis=0)
```

```python
df["bare_nuclei"] = df["bare_nuclei"].astype(int)
```

```python
df.dtypes
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(x,y)
```

```python
val_result = pd.read_csv(r"C:\Users\Lenovo\Desktop\Edu bridge\EDU\DATA\validation_results1.csv")

```

```python
val = pd.read_csv(r"C:\Users\Lenovo\Desktop\Edu bridge\EDU\DATA\breast_cancer_validation1.csv")
```

```python
y_pred = model.predict(val)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_val = val_result.iloc[:,-1]
```

```python
accuracy_score(y_val,y_pred)
```

```python
confusion_matrix(y_val,y_pred)
```
