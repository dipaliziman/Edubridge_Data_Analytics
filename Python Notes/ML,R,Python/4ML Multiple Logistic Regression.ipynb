## Check the loan approval status on the basis of salary

```python
df["ApplicantIncome"].describe()
```

```python
def salary(income):
    if income<2877:
        return "low"
    
    elif income<3812:
        return "medium"
    
    else:
        return "high"
    
```

```python
df["ApplicantIncome"]=df["ApplicantIncome"].apply(salary)
```

```python
df["ApplicantIncome"]
```

```python
df
```

```python
p=pd.crosstab(df["ApplicantIncome"],df["Loan_Status"])
```

```python
p
```

```python
p.plot(kind='bar',stacked=True)
plt.show()
```

if your income is high you get max loan 
if your income is low you get min loan
