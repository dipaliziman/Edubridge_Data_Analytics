```r
x<-c(1,2,3,4,5)
y<-c(6,7,8,9,10)
plot(x,y)
```

```r
labelset<-c("A","B","C","D","E")
plot(x,y,type='o',lty=3,col="blue",main="Data Visual",xlab="Data",ylab="Values")
text(x+0.5,y,labelset,col="red")
```

```r
x<-c(5,7,10)
y<-c(42,97,99)
x1<-c(6,8,9)
y1<-c(45,78,56)
```

```r
# scatter plot
plot(x,y,main="scatterplot",xlab="x-values",ylab="y-values")
points(x1,y1,col="red")
```

```r
# piechart
x<-c(7,5,4,6,2,9)
l<-c("A","B","C","D","E","F")
pie(x,labels<-l,init.angle=90)
```

```r
x<-c(7,5,4,6,2)
l<-c("A","B","C","D","E")
colors<-c("red","green","blue","pink","black")

pie(x,labels<-l,col=colors,init.angle=90)
legend("bottomright",l,fill=colors)
```

```r
# Bargraph 
x<-c("Delhi","Bangalore","Noida","mumbai")
y<-c(56,78,32,45)
colors<-c("red","green","blue","pink")
barplot(y,names.arg=x,col=colors)
```

```r
x<-c("Delhi","Bangalore","Noida","mumbai")
y<-c(56,78,32,45)
colors<-c("red","green","blue","pink")
barplot(y,names.arg=x,col=colors,density=20)
```
