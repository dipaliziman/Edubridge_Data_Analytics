```r
slices <- c(10, 12,4, 16, 8)
pct <-
round(slices/sum(slices)*100)
lbls <- paste(c("US", "UK", 
"Australia", 
"Germany", "France"), " ", pct, 
"%", sep="")
pie(slices, labels=lbls, 
col=rainbow(5),main="Pie Chart 
with Percentages")
```

```r
install.packages("plotrix")
```

```r
library(plotrix)
slices <- c(10, 12,4, 16, 8)
lbls <- paste(c("US", "UK", 
"France", 
"Germany", "Australia"), " ", pct, 
"%", sep="")
pie3D(slices,labels=lbls,explode=0.0,
main="3D Pie Chart")
```
